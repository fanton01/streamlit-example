````python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "ThresholdDataHeatMap.csv"  # Replace with your actual file path
df = pd.read_csv(file_path)

# Set the correlation threshold
# threshold_value = 0.75  # Replace with your desired threshold value

# Filter edges based on the correlation threshold
filtered_edges = df[['x', 'y']]

# Build a network using NetworkX
G = nx.from_pandas_edgelist(filtered_edges, 'x', 'y')

# Find connected components
connected_components = list(nx.connected_components(G))

# Extract the giant connected component
gcc_nodes = max(connected_components, key=len)
gcc = G.subgraph(gcc_nodes)
````

• The size (number of nodes), average degree, and average clustering in
the Giant Connected Component (GCC) you found in step (f).


````python
gcc_size = len(gcc_nodes)
gcc_avg_degree = sum(dict(gcc.degree()).values()) / gcc_size
gcc_avg_clustering = nx.average_clustering(gcc)
print("Giant Connected Component:")
print(f"Size: {gcc_size}")
print(f"Average Degree: {gcc_avg_degree}")
print(f"Average Clustering: {gcc_avg_clustering}")
````

````python
# Calculate log-log degree distribution, average degree of neighbors, and average clustering
degree_sequence_gcc = sorted([d for n, d in gcc.degree()], reverse=True)
degree_counts_gcc = pd.Series(degree_sequence_gcc).value_counts().sort_index()
#degree_sequence = list(dict(gcc.degree()).values())
avg_degree_neighbors = nx.average_degree_connectivity(gcc)
avg_clustering = nx.average_clustering(gcc)


plt.scatter(degree_counts_gcc.index, degree_counts_gcc.values, marker='o', color='skyblue')
plt.xscale('log')
plt.yscale('log')
plt.title('Degree Distribution')
plt.xlabel('Degree (k)')
plt.ylabel('P(k)')
plt.show()

x = list(avg_degree_neighbors.keys())
y = list(avg_degree_neighbors.values())
plt.scatter(x,y, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.title('Average Degree of Neighbors')
plt.xlabel('Degree (k)')
plt.ylabel('<k_nn(k)>')
plt.show()

x = list(nx.degree_centrality(gcc).values())
y = list(nx.clustering(gcc).values())
plt.scatter(x, y, marker='o')
plt.xscale('log')
plt.yscale('log')
plt.title('Average Clustering')
plt.xlabel('Degree (k)')
plt.ylabel('<C(k)>')
plt.show()

````

The plot of the distance (distribution), in normal scale

````python
# calculate distance distribution
all_distances = []
for node in gcc.nodes():
    distances = nx.shortest_path_length(gcc, source=node)
    all_distances.extend(list(distances.values()))

# Create a histogram of distances
plt.hist(distances, bins=20, density=True, edgecolor='black')
plt.title('Distribution of Distances in GCC')
plt.xlabel('Distance')
plt.ylabel('Frequency')
````
